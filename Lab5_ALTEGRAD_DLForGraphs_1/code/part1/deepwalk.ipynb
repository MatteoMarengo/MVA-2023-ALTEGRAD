```python
"""
Deep Learning on Graphs - ALTEGRAD - Nov 2023
MARENGO Matteo
matteo.marengo@ens-paris-saclay.fr
"""

import numpy as np
import networkx as nx
from random import randint
from gensim.models import Word2Vec


############## Task 1
# Simulates a random walk of length "walk_length" starting from node "node"
def random_walk(G, node, walk_length):

    ##################
    walk = [node] # walk starts from node
    for i in range(walk_length): # walk has length walk_length
        neighbour = list(G.neighbors(i)) # list of neighbors of current node
        walk.append
        walk.append(np.random.choice(list(G.neighbors(walk[-1])))) # next node is randomly chosen among neighbors of current node
    # permute walk
    walk = np.random.permutation(walk)
    ##################
    walk = [str(node) for node in walk]
    # what does this line does ?
    # it transforms the nodes from integers to strings
    # this is done because the Word2Vec model only accepts strings as input
    # this is done by the str(node) function
    return walk


############## Task 2
# Runs "num_walks" random walks from each node
def generate_walks(G, num_walks, walk_length):
    walks = []
    
    ##################
    # Runs "num_walks" random walks from each node
    for node in G.nodes():
        for i in range(num_walks):
            walks.append(random_walk(G, node, walk_length))
    # permute walks
    permuted_walks = np.random.permutation(walks)
    # describe what is a permutation
    # a permutation is a random shuffling of the walks
    # this is done to avoid that the walks are ordered by node
    # and that the model learns the nodes instead of the walks
    # this is done by the np.random.permutation function
    ##################

    return permuted_walks.tolist()


# Simulates walks and uses the Skipgram model to learn node representations
def deepwalk(G, num_walks, walk_length, n_dim):
    print("Generating walks")
    walks = generate_walks(G, num_walks, walk_length)

    print("Training word2vec")
    model = Word2Vec(vector_size=n_dim, window=8, min_count=0, sg=1, workers=8, hs=1)
    model.build_vocab(walks)
    model.train(walks, total_examples=model.corpus_count, epochs=5)

    return model
```
